## What is Python?

Python is a programming language — a way to give instructions to a computer. It's extremely popular in data science and Machine Learning (التعلم الآلي) because:
- It reads almost like English
- It has powerful libraries for math, data, and visualization
- It's free and runs in the cloud (like this notebook)

Let's start from the very beginning.

## 1. Variables and Data Types

A **variable** is just a name you give to a piece of information so you can use it later.
Think of it like a sticky note: you write a label on it and stick it to a value.

```python
# This is a comment — Python ignores everything after the # symbol

# Storing numbers
house_size = 2400      # integer (whole number)
price = 540000.0       # float (decimal number)

# Storing text (called a "string" in Python)
neighborhood = "Midtown"

# Storing True/False (called a "boolean")
has_pool = True

# Print them out
print("House size:", house_size, "sq ft")
print("Price: $", price)
print("Neighborhood:", neighborhood)
print("Has pool:", has_pool)
```

### Quick check
Run the cell above. You should see four lines of output.

Notice Python uses `=` to *assign* a value (not to say two things are equal). The `print()` function displays values on screen.

## 2. Lists — Storing Multiple Values

A **list** is an ordered collection of values. We use square brackets `[ ]`.

```python
# A list of house sizes (in square feet)
house_sizes = [1200, 1800, 2400, 3000, 3600]

# A list of prices (in $1000s)
prices = [220, 340, 500, 650, 800]

print("Sizes:", house_sizes)
print("First size:", house_sizes[0])    # indexing starts at 0!
print("Last size:",  house_sizes[-1])   # -1 gives the last item
print("Number of items:", len(house_sizes))
```

## 3. Loops — Doing Things Repeatedly

A **for loop** lets you repeat an action for each item in a list.

```python
# Print each house size
for size in house_sizes:
    print(f"House size: {size} sq ft")
```

```python
# You can also loop with an index using enumerate()
for i, size in enumerate(house_sizes):
    print(f"House {i+1}: {size} sq ft — costs ${prices[i]}k")
```

## 4. Functions — Reusable Instructions

A **function** is a block of code you name and can reuse. We define it with `def`.

```python
def predict_price(square_feet):
    """Simple rule-based price estimator."""
    price_per_sqft = 250   # $250 per sq ft
    base_price = 50000     # base cost
    return base_price + price_per_sqft * square_feet

# Use the function
print("Predicted price for 1500 sq ft: $", predict_price(1500))
print("Predicted price for 2500 sq ft: $", predict_price(2500))
print("Predicted price for 3500 sq ft: $", predict_price(3500))
```

Notice: we're doing exactly what we talked about in the first animation — using a formula with a weight (250) and a bias (50,000) to predict a price. That's a hand-written model. Machine Learning (التعلم الآلي) will *learn* those numbers automatically.

## 5. NumPy — Fast Math on Arrays

**NumPy** (Numerical Python) lets us do math on entire lists of numbers at once — much faster and cleaner than looping.

```python
import numpy as np   # 'np' is the conventional short name

# Create arrays (like lists, but much faster for math)
sizes  = np.array([1200, 1800, 2400, 3000, 3600])
prices = np.array([220,  340,  500,  650,  800])

print("Sizes:", sizes)
print("Average size:", np.mean(sizes))
print("Max price:   ", np.max(prices))
print("Std deviation of prices:", np.std(prices).round(1))
```

```python
# Math operations work on the WHOLE array at once
prices_in_dollars = prices * 1000   # multiply every element by 1000
print("Prices in dollars:", prices_in_dollars)

# Slicing: grab a subset
print("First three prices:", prices[:3])  # index 0, 1, 2
```

## 6. Pandas — Working with Tables

**Pandas** is the go-to library for working with tabular data (like Excel spreadsheets). A `DataFrame` is a table with labeled rows and columns.

```python
import pandas as pd  # 'pd' is the conventional short name

# Create a small dataset from scratch
data = {
    "size_sqft":  [1200, 1800, 2400, 3000, 3600],
    "bedrooms":   [2, 3, 4, 4, 5],
    "bathrooms":  [1, 2, 2, 3, 3],
    "price_k":    [220, 340, 500, 650, 800],
}

df = pd.DataFrame(data)
df   # Just typing the variable name in Jupyter shows a nice table
```

```python
# Exploring a DataFrame
print("Shape (rows, columns):", df.shape)
print()
print("Column names:", df.columns.tolist())
print()
print("First 3 rows:")
print(df.head(3))
print()
print("Basic statistics:")
print(df.describe().round(1))
```

```python
# Accessing columns
print("All sizes:", df["size_sqft"].values)
print()
# Filter rows: only houses bigger than 2000 sq ft
big_houses = df[df["size_sqft"] > 2000]
print("Big houses (>2000 sq ft):")
print(big_houses)
```

## 7. Matplotlib — Plotting

**Matplotlib** creates charts and graphs. Visualization is critical in Machine Learning (التعلم الآلي) for understanding your data.

```python
import matplotlib.pyplot as plt

# Scatter plot: size vs price
plt.figure(figsize=(8, 5))
plt.scatter(df["size_sqft"], df["price_k"], color="steelblue", s=100, zorder=5)

plt.xlabel("House Size (sq ft)", fontsize=13)
plt.ylabel("Price ($1000s)", fontsize=13)
plt.title("House Size vs. Price", fontsize=15)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Notice the pattern: as size increases, price increases. This is a positive correlation.")
```

```python
# Bar chart: average price by number of bedrooms
avg_by_beds = df.groupby("bedrooms")["price_k"].mean()

plt.figure(figsize=(7, 4))
avg_by_beds.plot(kind="bar", color="coral", edgecolor="black")
plt.xlabel("Number of Bedrooms", fontsize=13)
plt.ylabel("Average Price ($1000s)", fontsize=13)
plt.title("Average Price by Bedroom Count", fontsize=15)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
